```python
import numpy as np, numexpr

x = np.random.rand(1_000_000)
y= np.random.rand(1_000_000)

#Normal Numpy(creates temp arrays)
mask = (x > 0.5) & (y < 0.5)

#Numexpr(no big temp arrays)
mask_fast = numexpr.evaluates(" (x > 0.5) & (y < 0.5)")
```

```python
import pandas as pd
rng = np.random.RandomState(42)
df1, df2, df3, df4 = (pd.DataFrame(rng.rand(1000, 3)) for i in range(4))

#Normal way
res1 = df1 + df2 + df3 + df4

print(res1)

#Normal way
res2 =pd.eval("df1 + df2 + df3+ df4")

print(res2)
```
