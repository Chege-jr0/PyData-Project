```python
#How to visualise a continuous error measurement# define the model and draw some data
import numpy as np
import matplotlib.pyplot as plt
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF

#Define the model
model = lambda x: x * np.sin(x)
xdata = np.array([1, 3, 5, 6, 8])
ydata = model(xdata)

#Define kernel (RBF = squared exponential kernel)
kernel = RBF(length_scale=1.0)

#Create Gaussian Process Regressor
gp = GaussianProcessRegressor(kernel=kernel, alpha=1e-2)

#Fit to data
gp.fit(xdata[:, np.newaxis], ydata)

#Predict
xfit = np.linspace(0, 10, 1000)
yfit, sigma = gp.predict(xfit[:, np.newaxis], return_std=True)

#Plot
plt.figure(figsize=(8,5))
plt.plot(xdata, ydata, "ro", label="Data")
plt.plot(xfit, yfit, "b-", label="GP mean prediction")
plt.fill_between(xfit, yfit - 2*sigma, yfit + 2*sigma, color="blue", alpha=0.2, label="95% confidence interval")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#Create grid
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

#Contour plot
plt.contour(X, Y, Z, levels=10, cmap='viridis')  # Contour lines
plt.contourf(X, Y, Z, levels=10, cmap='viridis', alpha=0.5)  # Filled contours (density-like)
plt.colorbar(label='z')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Contour Plot of z = sin(sqrt(x² + y²))')
plt.show()
```
