## Pandas

Pandas is python library built on top of NumPy.

It provides high-perfomance data sctructures and tools for working with labeled data.

Core Objects:
Series => 1D labeled array
DataFrame => 2D labeled table(like excel/sql)

Pandas provides an efficient implementation of a DataFrame.

DataFrames are essentially multidimensional arrays with attched row and column labels and often heterogeneous types and/or missing data.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#A series is an array with labels(index)
s = pd.Series([10,20,30,40], index = ["a", "b", "c", "d"])
print(s)
```

```python
#A dataframe is like an excel or sql table
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25,30,35],
    "City": ["Nairobi", "Kisumu", "Mombasa"]
}
df = pd.DataFrame(data)
print(df)
```

```python
#Example on data merging
df1 = pd.DataFrame({'id': [1, 2, 3], 'value1': [10,20,30]})
df2 = pd.DataFrame({'id': [1, 2, 3], 'value1': [100,200,300]})

df_merged = pd.merge(df1, df2, on= 'id')

print(df_merged)
```

```python
#data indexers
data = pd.Series(['a', 'b', 'c'], index = [1,3,5])
print(data)
print("---------------------")
#print(data[1])
print(data.loc[1])
print("---------------------")
print(data.iloc[1])
print("---------------------")
#slicing with loc and iloc
print(data.loc[1:3])
print("---------------------")
print(data.iloc[1:3])
```

# Pandas string operations
Pandas provides a set of vectorized string operations via the .str attribute on Series and Index objects. This is crucial since the string operations can hanldle missing data.

```python
monte = pd.Series(['Graaham Chapman', 'John Cleese', 'Terry Gilliam' 'Eric Dundle', 'Terry Jones', 'Michael Palin'])
print(monte.str.capitalize())
print("-----------------")
print(monte.str.len())
print("-----------------")
monte.str.lower()
print("------------------------")
monte.str.startswith("T")
print("---------------------")
monte.str.split
```

# USING REGEX to operate on strings

```python
monte.str.extract('([A-Za-z]+)', expand=False)
```

```python
monte.str.findall(r'^[^AEIOU].*[aeiou]$')
```

```python
monte.str.replace('Terry', 'Terrence')
```

```python
monte.str.count('J')
```
